## Problem: Allocating Labor between Two Tasks
We have a total labor constraint $\bar{L} = 1$, where $L_1 + L_2 = 1$.
The production functions for each task are:

$Q_1 = L_1^{0.5}$
$Q_2 = L_2^{0.5}$

By substituting $L_2 = 1 - L_1$, we can express total output as a function of $L_1$:
$Q_{Total} = L_1^{0.5} + (1 - L_1)^{0.5}$

## Allocating $L_1$ and $L_2$

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup Data
a = 10
L1 = np.linspace(0, a, 500)
Q1 = L1**0.5
Q2_mapped = (a - L1)**0.5
Total_Q = Q1 + Q2_mapped

# 2. Create the Figure
# height_ratios=[1, 1, 2.5] makes the 3rd row much taller than the 1st and 2nd
fig = plt.figure(figsize=(14, 22))
gs = fig.add_gridspec(3, 2, hspace=0.4, wspace=0.4, height_ratios=[1, 1, 2.5])

# --- ROW 1: Base Functions ---
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(L1, Q1, color='blue', lw=2)
ax1.set_title(r'1: $Q_1 = L_1^{0.5}$', fontsize=12)
ax1.set_xlabel('$L_1$')
ax1.grid(alpha=0.3)

ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(L1, Q1, color='green', lw=2)
ax2.set_title(r'2: $Q_2 = L_2^{0.5}$', fontsize=12)
ax2.set_xlabel('$L_2$')
ax2.grid(alpha=0.3)

# --- ROW 2: Mapping to L1 ---
ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(L1, Q1, color='blue', lw=2)
ax3.set_title(r'3: $Q_1 = L_1^{0.5}$', fontsize=12)
ax3.set_xlabel('$L_1$')
ax3.grid(alpha=0.3)

ax4 = fig.add_subplot(gs[1, 1])
ax4.plot(L1, Q2_mapped, color='green', lw=2)
ax4.set_title(r'4: $Q_2 = (a - L_1)^{0.5}$', fontsize=12)
ax4.set_xlabel('$L_1$')
ax4.grid(alpha=0.3)

# --- ROW 3: The Vertical Summation (LONG/TALL FIGURE) ---
ax5 = fig.add_subplot(gs[2, 0])
ax5.plot(L1, Q1, 'b--', alpha=0.3, label='$Q_1$ component')
ax5.plot(L1, Q2_mapped, 'g--', alpha=0.3, label='$Q_2$ component')
ax5.plot(L1, Total_Q, color='red', lw=3, label='Sum ($Q_1 + Q_2$)')

ax5.set_title('5: Vertical Summation (Long View)', fontsize=14)
ax5.set_xlabel('$L_1$', fontsize=12)
ax5.set_ylabel('Total Output', fontsize=12)
ax5.set_ylim(0, a)
ax5.grid(alpha=0.3)

# Legend moved outside to the right
ax5.legend(loc='upper left', bbox_to_anchor=(1.05, 1), borderaxespad=0.)

plt.show()
```

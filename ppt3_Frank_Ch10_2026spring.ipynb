## The Goal of Profit Maximization

- Suppose a firm produces 100 units of output per week.
- Using 10 units of $L$ and $K$, respectively.
- Price of labor ( $w$, wage) and price of capital ( $r$, interest rate) is \$10/unit, respectively
- Price of the output $(p)$ is $\$ 2.5 /$ unit
- Total revenue will be $\$ 250 / \mathrm{wk}$
- Say that 5 units of capital is owned by the firm



---

## The Goal of Profit Maximization

- Economic profit is the difference between total revenue and total cost, where total cost includes all costs-both explicit and implicit-associated with resources used by the firm.
- Accounting profit is simply total revenue less all explicit costs incurred.
- Does not subtract the implicit costs.
- Economists assume that the goal of firms is to maximize economic profit.

## SELF-TEST 10.1

Suppose a firm has $\$ 250$ of weekly revenue, $\$ 100$ of weekly labor cost, and $\$ 150$ of implicit weekly capital cost. What would be the firm's economic and accounting profit for the week?
□
Answer

The firm's economic profit would be $\$ 0$, while its accounting profit would remain at $\$ 150$. As will be seen, zero economic profits is the theoretically expected outcome under perfect competition.

## The Four Conditions for Perfect Competition

1. Firms sell a standardized product.

- The product sold by one firm is assumed to be a perfect substitute for the product sold by any other.

2. Firms are price takers.

- The individual firm treats the market price of the product as given.

3. Free entry and exit, with perfectly mobile factors of production in the long run.

- If a firm perceives an opportunity, it can hire the factors of production required to take advantage of it.

4. Firms and consumers have perfect information.

## The Short-Run Condition for Profit Maximization

- Total revenue:

$$
T R=P \times Q
$$

- Marginal revenue: the change in total revenue that occurs as a result of a 1-unit change in sales.

$$
M R=\frac{\Delta T R}{\Delta Q}=P
$$


---

## The Short-Run Condition for Profit Maximization

- Profit

$$
\text { Profit }=T R-T C
$$

- To maximize profits the firm should produce a level of output for which marginal revenue is equal to marginal cost on the rising portion of the MC curve.

## The Short-Run Competitive Industry Supply

- The short-run supply curve for a competitive industry is generated by announcing a price and then adding together the amounts each firm wishes to supply at that price.
- The resulting sum is industry supply at that price.

## Short-Run Competitive Equilibrium

- Even though the market demand curve is downward sloping, the demand curve facing the individual firm is perfectly elastic.
- The breakeven point is the point at which price is equal to the minimum of average total cost.
- The lowest price at which the firm will not suffer negative profits in the short run.

## The Efficiency of Short-Run Competitive Equilibrium

- One of the most attractive features of competitive markets is the fact that they result in allocative efficiency.
- Allocative efficiency: a condition in which all possible gains from exchange are realized.
- At the short-run competitive equilibrium price and quantity, the value of the resources used to produce the last unit of output (as measured by short-run marginal cost) is exactly equal to the value of that unit of output to consumers (as measured by the price they are willing to pay for it).

## Producer Surplus

- A competitive market is efficient when it maximizes the net benefits to its participants.
- Producer surplus: the dollar amount by which a firm benefits by producing a profit-maximizing level of output.

# Lecture: Aggregate Marginal Cost for Multi-Plant Production

In a multi-plant firm, minimizing total cost requires allocating production such that the **Marginal Cost (MC)** is equal across all sites. This process is known as **Horizontal Summation**.

## 1. Define Individual Marginal Costs
Marginal cost is the first derivative of the Total Cost ($TC$) function.

* **Site A:** $TC_A = 16 + 6Q_A^2$
    
    $$MC_A = \frac{dTC_A}{dQ_A} = 12Q_A$$

* **Site B:** $TC_B = 240 + 2Q_B^2$
    
    $$MC_B = \frac{dTC_B}{dQ_B} = 4Q_B$$

---

## 2. Horizontal Summation Logic
To find the aggregate curve, we sum the **quantities** ($Q$) at each level of marginal cost ($MC$).

### Step A: Invert the functions
Express Quantity ($Q$) as a function of $MC$:
* **Site A:** $Q_A = \frac{MC}{12}$
* **Site B:** $Q_B = \frac{MC}{4}$

### Step B: Aggregate Quantities
Sum the individual quantities to find the total quantity ($Q_T$):
$$Q_T = Q_A + Q_B = \frac{MC}{12} + \frac{MC}{4}$$

### Step C: Simplify and Re-invert
Find a common denominator to solve for $Q_T$:

$$Q_T = \frac{1MC}{12} + \frac{3MC}{12} = \frac{4MC}{12} = \frac{MC}{3}$$

**Aggregate Marginal Cost Function:**

$$MC_{agg} = 3Q_T$$

---

## 3. Key Summary
| Metric | Site A | Site B | Aggregate |
| :--- | :--- | :--- | :--- |
| **Total Cost ($TC$)** | $16 + 6Q_A^2$ | $240 + 2Q_B^2$ | $TC_{min}$ |
| **Marginal Cost ($MC$)** | $12Q_A$ | $4Q_B$ | $3Q_T$ |
| **Inverse ($Q$)** | $MC/12$ | $MC/4$ | $MC/3$ |

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# 1. Symbolic Calculation
Qa, Qb, MC = sp.symbols('Qa Qb MC')

# Define Total Cost functions
TC_A = 16 + 6*Qa**2
TC_B = 240 + 2*Qb**2

# Calculate Marginal Costs
MC_A_func = sp.diff(TC_A, Qa)
MC_B_func = sp.diff(TC_B, Qb)

print(f"MC_A: {MC_A_func}")
print(f"MC_B: {MC_B_func}")

# 2. Visualization
q_values = np.linspace(0, 100, 400)

# MC_A = 12*Q -> MC/12 = Q
# MC_B = 4*Q  -> MC/4 = Q
# MC_agg = 3*Q -> MC/3 = Q

def mc_a(q): return 12 * q
def mc_b(q): return 4 * q
def mc_agg(q): return 3 * q

plt.figure(figsize=(10, 6))
plt.plot(q_values, mc_a(q_values), label='MC_A (12Q)', linestyle='--')
plt.plot(q_values, mc_b(q_values), label='MC_B (4Q)', linestyle='--')
plt.plot(q_values, mc_agg(q_values), label='Aggregate MC (3Q)', linewidth=3, color='black')

plt.title('Horizontal Summation of Marginal Costs')
plt.xlabel('Quantity (Q)')
plt.ylabel('Marginal Cost ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 50)
plt.ylim(0, 200)
plt.show()
```
